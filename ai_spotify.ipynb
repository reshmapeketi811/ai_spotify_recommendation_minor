```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("spotify dataset.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()

# Fill numeric columns with median
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
df['track_album_release_date'] = pd.to_datetime(
    df['track_album_release_date'], errors='coerce'
)
```

```python
df.drop_duplicates(subset='track_id', inplace=True)
```

```python
plt.figure()
sns.histplot(df['track_popularity'], bins=30)
plt.title("Track Popularity Distribution")
plt.show()
```

```python
plt.figure()
df['playlist_genre'].value_counts().plot(kind='bar')
plt.title("Songs per Genre")
plt.show()
```

```python
sns.pairplot(df[['danceability','energy','valence','tempo']])
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df[num_cols].corr()
sns.heatmap(corr, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
features = [
    'danceability','energy','loudness','speechiness',
    'acousticness','instrumentalness','liveness',
    'valence','tempo'
]

X = df[features]

```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
plt.figure()
sns.scatterplot(
    x=df['danceability'],
    y=df['energy'],
    hue=df['cluster'],
    palette='tab10'
)
plt.title("Song Clusters")
plt.show()
```

```python
plt.figure()
sns.countplot(x='cluster', hue='playlist_genre', data=df)
plt.title("Clusters by Genre")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='cluster', hue='playlist_name', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
similarity = cosine_similarity(X_scaled)
```

```python
def recommend_song(song_name, df, similarity, n=5):
    idx = df[df['track_name'] == song_name].index[0]
    scores = list(enumerate(similarity[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    scores = scores[1:n+1]

    song_indices = [i[0] for i in scores]
    return df[['track_name','track_artist']].iloc[song_indices]
```

```python
recommend_song("Memories - Dillon Francis Remix", df, similarity)
```
